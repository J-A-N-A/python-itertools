# ***ITERTOOLS PYTHON ***


# *infinite iterators*

```python
#itertools count
from itertools import count
x=(count(start=0,step=2))
print("prints all even numbers of range 10",list(next(x) for _ in range(10)))   #The next() function returns the next item from the iterator
```

```python
#another example
from itertools import count
list = ['John', 'Marie', 'Jack', 'Anna','Marie']
for i in zip(count(), list):  #The zip() function takes iterables (can be zero or more), aggregates them in a tuple, and returns it.
    print(i)
```

```python
#itertools cycle      ##This process is repeated indefinitely!!!!!.  u may run out of memory careful
from itertools import cycle
lis=[1,2,3,4]
x=(cycle(lis))
for i in x:print(i)
```

```python
#itertools repeat
from itertools import repeat
x=repeat(10,3)
print("repeats 10 three times",i)
for i in x:print(i)
```

# *Iterators terminating on shortest input sequence*

***ACCUMULATE*** <BR>
Syntax
itertools.accumulate(iterable[, func]) > accumulate object

```python
from itertools import accumulate
import operator    #operator module
lis=[1,2,3,45,6,78,9]
res=accumulate(lis,operator.mul)
print("this is actually multiplying the pair of list elements")
for i in res:print(i)
```

```python
#another example
from itertools import accumulate
import operator
lis=[2,3,4,5,6,77,8,9,9,0]
res=accumulate(lis,max)
print("this is actually comapring the 2 pairs of iterators and printing the max val")
for i in res:print(i)
```

# **combinatoric iterators**

# *product()*

```python
#product()

from itertools import product
x=product('ABCD',repeat=2)
for i in x:print(i)
```

# *permutations()*
A permutation is a mathematical technique that determines the number of possible arrangements in a set when the order of the arrangements matters.

```python
from itertools import permutations
lis=[1,2,3,4,5]
x=permutations(lis,2)
for i in x:print(i)
```

# *combinations()*
It is the different selections of a given number of elements taken one by one, or some, or all at a time

```python
from itertools import combinations
lis=[1,2,3]
x=combinations(lis,2)
for i in x:print(i)
```

# *combinations_with_replacement()*
combinations_with_replacement means all the possible arrangements or subsets that allow an element to repeat in a subset.

```python
from itertools import combinations_with_replacement
lis=[1,2,3]
x=combinations_with_replacement(lis,2)
for i in x:print(i)
```
